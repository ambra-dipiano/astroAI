```python
import matplotlib.pyplot as plt

plt.hist([dsg, dsm], bins=50, histtype='step', color=['b', 'r'])
plt.xlim([-0.5, 2.5])
plt.axvline(np.median(dsg), c='b')
plt.axvline(np.median(dsm), c='r')
plt.show()

```
